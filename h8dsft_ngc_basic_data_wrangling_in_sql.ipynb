```python
# put your query here

query = """
SELECT COUNT(*) AS table_count
FROM`bigquery-public-data.austin_bikeshare.bikeshare_trips`
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

```python
query2 = """
SELECT COUNT('name') AS string_column_count
FROM `bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS`
WHERE table_name = 'bikeshare_stations' AND data_type = 'STRING';

"""
# Set up the query
query_job2 = client.query(query2).to_dataframe()
query_job2
```

```python
query3 = """
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE bike_id IS NOT NULL
  AND duration_minutes > 1000;
"""
# Set up the query
query_job3 = client.query(query3).to_dataframe()
query_job3
```
